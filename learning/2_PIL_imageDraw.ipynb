# 使用PIL库来添加标签看一下效果如何

```python
from PIL import Image, ImageDraw, ImageFont

img = Image.open("image/dog.jpg")
draw = ImageDraw.Draw(img)
font = ImageFont.truetype("font/simhei.ttf", 20)
```

```python
img
```

```python
img.size, img.mode, img.format
```

```python
# 计算bounding box的左上角和右下角坐标
x1, y1, x2, y2 = 50, 50, 100, 150

# 绘制bounding box
draw.rectangle((x1, y1, x2, y2), outline='red', width=1)

img
```

```python
# 添加文字标签
text = '萨摩耶'
x, y = 50, 50 # 标签左上角的坐标
draw.text(
    (x, y), text, fill='red', font=font
)

img
```
